## Challenge 1: Combining & Cleaning Data

In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv

But this time, we got new data, which can be found in the following 2 CSV files located at the links below.

- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv

Note that you'll need to clean and format the new data.

Observation:
- One option is to first combine the three datasets and then apply the cleaning function to the new combined dataset
- Another option would be to read the clean file you saved in the previous lab, and just clean the two new files and concatenate the three clean datasets

```python
# 1

# Import pandas
import pandas as pd

# Step 1: Load all three datasets
file1_url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
file2_url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
file3_url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

# Load the datasets
file1 = pd.read_csv(file1_url)
file2 = pd.read_csv(file2_url)
file3 = pd.read_csv(file3_url)

# Step 2: Define a cleaning function for consistent formatting
def clean_data(df):
    # Standardize column names
    df.columns = (
        df.columns.str.lower()          # Convert to lowercase
              .str.replace(" ", "_")    # Replace spaces with underscores
              .str.replace("st", "state")  # Replace 'st' with 'state'
    )
    
    # Clean and standardize the Gender column
    if "gender" in df.columns:
        df["gender"] = df["gender"].str.strip().str.lower().replace({
            "f": "F", 
            "female": "F", 
            "m": "M", 
            "male": "M", 
            "femal": "F"
        })
    
    # Replace state abbreviations with full names
    state_mapping = {
        "AZ": "Arizona",
        "Cali": "California",
        "WA": "Washington"
    }
    if "state" in df.columns:
        df["state"] = df["state"].replace(state_mapping)
    
    # Replace "Bachelors" with "Bachelor" in the Education column
    if "education" in df.columns:
        df["education"] = df["education"].str.strip().str.replace("Bachelors", "Bachelor", regex=False)
    
    # Remove the % character from Customer Lifetime Value
    if "customer_lifetime_value" in df.columns:
        df["customer_lifetime_value"] = (
            df["customer_lifetime_value"]
            .astype(str)
            .str.replace("%", "", regex=False)
            .astype(float)
        )
    
    # Standardize the Vehicle Class column
    vehicle_class_mapping = {
        "Sports Car": "Luxury",
        "Luxury SUV": "Luxury",
        "Luxury Car": "Luxury"
    }
    if "vehicle_class" in df.columns:
        df["vehicle_class"] = df["vehicle_class"].replace(vehicle_class_mapping)
    
    return df

# Step 3: Apply cleaning function to all three datasets
file1_cleaned = clean_data(file1)
file2_cleaned = clean_data(file2)
file3_cleaned = clean_data(file3)

# Step 4: Combine the cleaned datasets
combined_data = pd.concat([file1_cleaned, file2_cleaned, file3_cleaned], ignore_index=True)

# Step 5: Handle duplicates and reset the index
combined_data = combined_data.drop_duplicates().reset_index(drop=True)

# Output basic information for verification
print("Combined dataset saved to 'combined_cleaned_data.csv'")
print(f"Shape of the combined dataset: {combined_data.shape}")
print("Preview of the combined dataset:")
print(combined_data.head())
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
# 2

# Step 1: Load the dataset
file_path = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv" 
marketing_data = pd.read_csv(file_path)

# Step 2: Drop unnecessary columns
marketing_data.drop(columns=["unnamed:_0"], inplace=True)

# Step 3: Standardize column names
marketing_data.columns = (
    marketing_data.columns.str.lower()  # Convert to lowercase
                  .str.replace(" ", "_")  # Replace spaces with underscores
)

# Step 4: Convert data types
# Convert 'effective_to_date' to datetime
marketing_data["effective_to_date"] = pd.to_datetime(marketing_data["effective_to_date"])

# Convert numeric columns to integers where applicable
numeric_columns = ["customer_lifetime_value", "months_since_last_claim", "number_of_open_complaints", "total_claim_amount"]
for col in numeric_columns:
    marketing_data[col] = marketing_data[col].astype(int)

# Step 5: Standardize categorical values
# Standardize 'gender'
marketing_data["gender"] = marketing_data["gender"].str.strip().str.lower().replace({
    "f": "F", "female": "F", "m": "M", "male": "M", "femal": "F"
})

# Standardize 'vehicle_class'
vehicle_class_mapping = {
    "Sports Car": "Luxury",
    "Luxury SUV": "Luxury",
    "Luxury Car": "Luxury"
}
marketing_data["vehicle_class"] = marketing_data["vehicle_class"].replace(vehicle_class_mapping)

# Standardize 'education'
marketing_data["education"] = marketing_data["education"].str.strip().str.replace("Bachelors", "Bachelor", regex=False)

marketing_data
```
